#### Data augmentation

Data augmentation is a process of creating new or additional data samples, which can include images, through various techniques such as rotation, flipping, or adding noise.

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Create an ImageDataGenerator object with some augmentation settings
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
# Reshape the image to a 4D tensor with a single sample
img = img.reshape((1,) + img.shape)
# Generate batches of augmented images
aug_iter = datagen.flow(img)
# Generate 9 augmented images
fig, axs = plt.subplots(3, 3, figsize=(10, 10))
for i in range(9):
    # Generate an augmented image
    augmented_img = aug_iter.next()[0].astype('uint8')
    axs[i//3, i%3].imshow(augmented_img)
plt.show()
```
